## a. Numerical vs Numerical

### b. Categorical vs Numerical
### Compare average salaries by Company or Gender:

```python
custom_palette = {
    'TCS': '#66c2a5',         
    'Infosys': '#fc8d62',     
    'Cognizant': '#8da0cb'    
}

# Create the boxplot
sns.boxplot(x='Company', y='Employee Salary', data=df, palette=custom_palette)
plt.title('Salary Distribution by Company')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
gender_palette = {'Male': 'skyblue', 'Female': 'lightpink'}

# Barplot: Average Salary by Gender
sns.barplot(x='Gender', y='Employee Salary', data=df, palette=gender_palette)
plt.title('Average Salary by Gender')
plt.tight_layout()
plt.show()
```

_______________
# 2. Check for Patterns or Trends
____________
## a. Salary by Location
____________

```python
# Average Salary by Place
avg_salary_by_place = df.groupby('Place')['Employee Salary'].mean().sort_values(ascending=False)
avg_salary_by_place.plot(kind='bar', figsize=(10, 4), title='Average Salary by Place', color='orange')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

__________
## b. Age Group Analysis
____________

```python
# Define bins and labels
bins = [18, 25, 35, 45, 60]
labels = ['18â€“25', '26â€“35', '36â€“45', '46â€“60']

# Create Age Group column
df['Age Group'] = pd.cut(df['Employee Age'], bins=bins, labels=labels)

# Define custom palette
age_palette = {
    '18â€“25': '#FF9999',   
    '26â€“35': '#66B2FF',   
    '36â€“45': '#99FF99',   
    '46â€“60': '#FFCC99'    
}

# Plot average salary by age group
plt.figure(figsize=(8, 5))
sns.barplot(
    x='Age Group', 
    y='Employee Salary', 
    data=df, 
    estimator='mean', 
    ci=None, 
    palette=age_palette
)
plt.title('Average Salary by Age Group')
plt.ylabel('Average Salary')
plt.xlabel('Age Group')
plt.tight_layout()
plt.show()
```

_____________
# 1. Pie Charts for Categorical Features
__________

```python
#Pie charts work well to understand proportions in smaller category sets (like Gender, Country):
categorical_cols = ['Company', 'Place', 'Country', 'Gender']

for col in categorical_cols:
    plt.figure(figsize=(6, 6))
    df[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.ylabel('')  
    plt.show()
```

_____________
# 2. Grouped Count Plots: Gender Split by Company
__________

```python
#This shows how many males/females are in each company:
plt.figure(figsize=(8, 5))
sns.countplot(x='Company', hue='Gender', data=df, palette='Set2')
plt.title('Gender Distribution Across Companies', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
